```python
##DEPENDENCIES##

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import ipyvolume as ipv
```

```python
##VARIABLES FOR PARALLEL CAMERAS##

focal_length = 1.93 #mm 
baseline = 50 #mm #distance between two parallel cameras 
ratio = 1.5 #ratio of pixels to mm
```

```python
#df1 is trajectory data from the first camera 
#df2 is trajectory data from the second camera 

#read and extract values 
df1 = pd.read_csv('')
df2 = pd.read_csv('')

ir1x = df1['X'].values 
ir1y = df1['Y'].values

ir2x = df2['X'].values 
ir2y = df2['Y'].values

#get disparity
ydisparity = abs(ir1y - ir2y) 
xdisparity = abs(ir1x -ir2x)

#calculate x,y,z 
Y_tmp = [y*baseline for y in ir1y]
X_tmp = [x*baseline for x in ir1x]

x = X_tmp/xdisparity
y = Y_tmp/ydisparity
z = [(focal_length*baseline) / (i/ratio) for i in ydisparity]

#visualize 
ipv.quickscatter(x, y, np.asarray(z), size=1, marker="sphere")
```
